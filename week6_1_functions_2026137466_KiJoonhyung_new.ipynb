```python
# First function example: Add 1 to a and store as b, return 반환값

def add(a):
    b = a + 1
    print(a, "if you add one", b)
    return(a, b) # y=x+1, (x,y)
add(1) #함수를 부른 것임 (call functions)
```

```python
# Get a help on add function, help 기능

help(add)
```

```python
# Define a function for multiple two numbers

def Mult(a, b):
    c = a * b
    return(c)
# Use mult() multiply two integers

result_mult = Mult(2.3, 3.1)
type(result_mult)
```

```python
# Use mult() multiply two floats
Mult(10.0, 3.14)
```

```python
# Function Definition, square 제곱함수 만들기
def square(a):

    # Local variable b
    b = 1
    c = a * a + b
    print(a, "if you square + 1", c)
    return(c)
    
# Initializes Global variable

x = 3
# Makes function call and return function a y
y = square(x)
y
```

```python
# Directly enter a number as parameter

square(2)
```

```python
# Example of global variable

artist = "Michael Jackson" #전역변수 = global...
def printer1(artist):
    internal_var = artist #local variable, 즉 printer1이 끝나면 없어지는 것임
    print(artist, "is an artist")

printer1(artist, internal_var) #artist는 알아, internal_var 몰라...
```

```python
artist = "Michael Jackson"

def printer(artist):
    global internal_var #일부러 internal_var이 전역이라는 것을 알려줄 수도 있다. --> 잘 안씀...
    internal_var= "Whitney Houston"
    print(artist,"is an artist")

printer(artist)
printer(internal_var)
```
