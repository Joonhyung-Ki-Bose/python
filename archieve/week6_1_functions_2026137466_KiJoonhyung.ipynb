```python
# Function Definition

def square(a):

    # Local variable b
    b = 1
    c = a * a + b
    print(a, "if you square + 1", c)
    return(c)
```

```python
# Initializes Global variable

x = 3
# Makes function call and return function a y
y = square(x)
y
```

```python
# Directly enter a number as parameter

square(2)
```

```python
# Example of global variable

myFavouriteBand = "AC/DC" #global

def getBandRating(bandname):
    if bandname == myFavouriteBand:
        return 10.0
    else:
        return 0.0

print("AC/DC's rating is:", getBandRating("AC/DC"))
print("Deep Purple's rating is:",getBandRating("Deep Purple"))
print("My favourite band is:", myFavouriteBand)
```

```python
def getBandRating(bandname):
    myFavouriteBand2 = "AC/DC" #local variable
    print("Inside getBandRating, myFavouriteBand2 is:", myFavouriteBand2)
    if bandname == myFavouriteBand2:
        return 10.0
    else:
        return 0.0

print("AC/DC's rating is: ", getBandRating("AC/DC"))
print("Deep Purple's rating is: ", getBandRating("Deep Purple"))
# The following line will still cause a NameError because myFavouriteBand2 is a local variable
# print("My favourite band is", myFavouriteBand2)
```

```python
# Example of global variable and local variable with the same name

myFavouriteBand = "AC/DC"

def getBandRating(bandname):
    myFavouriteBand = "Deep Purple"
    if bandname == myFavouriteBand:
        return 10.0
    else:
        return 0.0

print("AC/DC's rating is:",getBandRating("AC/DC"))
print("Deep Purple's rating is: ",getBandRating("Deep Purple"))
print("My favourite band is:",myFavouriteBand)
```
