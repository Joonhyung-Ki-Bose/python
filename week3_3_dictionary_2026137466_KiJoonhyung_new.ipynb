# Dictionaries in Python

```python
# Create the dictionary, Dictionary는 key값을 통해 데이터에 접근, 데이터베이스에 특화된 데이터 타입, 중괄호 사용

Dict = {"key1": 1, "key2": "2", "key3": [3, 3, 3], "key4": (4, 4, 4), 'key5': 5, (0, 1): 6, 1.2: 5}
#{} :
# key1 (키값) --> 1 (정수)
# key2 (키값) --> "2" (문자열)
# key4 (키값) --> (4,4,4) tuple
# (0,1) tuple 키값 --> 5 (정수), 리스트[]는 키캆이 될수 없다?
Dict
```

```python
# Access to the value by the key
#Dict[1.2]
#Dict['key5']
#Dict[(0,1)]
Dict["key3"]
```

```python
# Access to the value by the key

Dict["key3"]="TEST" #다른 값을 넣을 수 있니?
Dict
```

```python
# Create a sample dictionary

release_year_dict = {"Thriller": "1982", "Back in Black": "1980", "The Dark Side of the Moon": "1973", "The Bodyguard": "1992", \
                    "Bat Out of Hell": "1977", "Their Greatest Hits (1971-1975)": "1976", \
                    "Saturday Night Fever": "1977", "Rumours": "1977"}
release_year_dict
```

```python
# Get value by keys

release_year_dict['Their Greatest Hits (1971-1975)']
```

```python
# Get all the keys in dictionary

release_year_dict.keys()
```

```python
# Get all the values in dictionary

release_year_dict.values()
```

```python
# Append value with key into dictionary

release_year_dict['Graduation'] = '2007'
release_year_dict
```

```python
# Delete entries by key

del(release_year_dict['Thriller'])
del(release_year_dict['Graduation'])
release_year_dict
```

```python
# We can verify if an element is in the dictionary:
'Thriller' in release_year_dict
```
