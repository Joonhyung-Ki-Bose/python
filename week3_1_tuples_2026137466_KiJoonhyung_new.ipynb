# Tuples in Python

```python
# Create your first tuple, Tuples : 파이썬의 datatype

tuple1 = ("disco",10,1.2 ) #tuple을 선언하는 방법은 ()
tuple1
```

```python
# Print the type of the tuple you created

type(tuple1)
```

```python
# Print the variable on each index, 각 요소에 접근

print(tuple1[0]) #"disco"
print(tuple1[1]) #10
print(tuple1[2]) #1,2
```

```python
# Use negative index to get the value of the last element

tuple1[-1]
```

```python
# Concatenate two tuples

tuple2 = tuple1 + ("hard rock", 10)
tuple2
```

```python
len(tuple2)
```

```python
# Slice from index 0 to index 2, 정렬

tuple2[0:5]
```

```python
# A sample tuple

Ratings = (0, 9, 6, 5, 10, 8, 9, 6, 2)
```

```python
# Sort the tuple

RatingsSorted = sorted(Ratings) #정렬...sorted
RatingsSorted
```

```python
# Create a nest tuple

NestedT =(1, 2, ("pop", "rock") ,(3,4),("disco",(1,2)))

#NestedT[0] --> 1
#NestedT[2] ---> ("pop", "rock") : tuple ()
#NestedT[3] ---> (3,4) : tuple ()
#NestedT[4] --->("disco", (1,2)) : tuple
#NestedT[4][0] -->"disco" :""
#NestedT[4][1] --> (1,2) : () tuple
#NestedT[4][1][0] --> 1 : integer
```

```python
# Print element on each index

print("Element 0 of Tuple: ", NestedT[0])
print("Element 1 of Tuple: ", NestedT[1])
print("Element 2 of Tuple: ", NestedT[2])
print("Element 3 of Tuple: ", NestedT[3])
print("Element 4 of Tuple: ", NestedT[4])
```

```python
# Print the first element in the second nested tuples

NestedT[2][1][0]
```
