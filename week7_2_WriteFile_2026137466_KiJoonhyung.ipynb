```python
# Sample list of text, 3줄의 문자열 생성

Lines = ["This is line A\n", "This is line B\n", "This is line C\n"]
Lines
```

```python
# Write the strings in the list to text file, 3줄의 문자열 생성을 for 루프를 이용하여 example11을 생성

with open("C:/Users/jk1050404/OneDrive - Bose Corporation/_MOT/파이썬프로그래밍/실습자료/example2.txt", 'w') as writefile:
    for line in Lines:
        print(line)
        writefile.write(line)
```

```python
# Verify if writing to file is successfully executed

with open("C:/Users/jk1050404/OneDrive - Bose Corporation/_MOT/파이썬프로그래밍/실습자료/example2.txt", 'r') as testwritefile:
    print(testwritefile.read())
```
