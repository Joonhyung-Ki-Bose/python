```python
# List comprehension
number_list = []
number_list.append(1)
number_list.append(2)
number_list.append(3)
number_list.append(4)
number_list.append(5)
number_list
```

```python
number_list = []
for number in range(1, 6):
    number_list.append(number)
number_list
```

```python
number_list = list(range(1, 6))
number_list
```

```python
number_list = [number for number in range(1,6)]
number_list
```

```python
number_list = [number-1 for number in range(1,6)]
number_list
```

```python
a_list = [number for number in range(1,6) if number % 2 == 1]
a_list
```

```python
a_list = []
for number in range(1,6):
    if number%2 == 1:
        a_list.append(number)
a_list
```

```python
sentence = ['I', 'Love', 'Python', 'Soooooo', 'MUCH!!!']
sentence
```

```python
#함수적용
[word.lower() for word in sentence]
```

```python
#조건적용
[word for word in sentence if len(word) > 6]
```

```python
#함수적용하여튜플로저장, 제곱, 삼승
[(x, x ** 2, x ** 3) for x in range(10)]
```

```python
rows = range(1,4)
cols = range(1,3)
for row in rows:
    for col in cols:
        print(row, col)
```

```python
rows = range(1,4)
cols = range(1,3)
cells = [(row, col) for row in rows for col in cols]
for cell in cells:
    print(cell)
```

```python
[(i,j) for i in range(5) for j in range(i)]
```

```python
# Dictionary Comprehension
word = 'letters'
letter_counts= {letter: word.count(letter) for letter in word}
letter_counts
```

```python
word = 'letters'
letter_counts= {letter: word.count(letter) for letter in set(word)}
letter_counts
```

```python
# Set Comprehension
a_set= {number for number in range(1,6) if number % 3 == 1}
a_set
```

```python
def sum(a, b):
    return a+b
print(sum(1.3, 3.1))
print(sum('love', 'python'))
```

```python
# iteration - zip
days = ['Monday', 'Tuesday', 'Wednesday']
fruits = ['banana', 'orange', 'peach']
drinks = ['coffee', 'tea', 'beer']
desserts = ['tiramisu', 'ice cream', 'pie', 'pudding']

for day, fruit, drink, dessert in zip(days, fruits, drinks, desserts):
    print(day, ": drink", drink, "-eat", fruit, "-enjoy", dessert)
```

```python
english = 'Monday', 'Tuesday', 'Wednesday'
french= 'Lundi', 'Mardi', 'Mercredi'
print(list(zip(english, french)))
print(dict(zip(english, french)))
```

```python
# 함수 위치 인자
def menu(wine, entree, dessert):
    return {'wine': wine, 'entree': entree, 'dessert': dessert}
menu('chardonnay', 'chicken', 'cake')
```

```python
# 함수 키워드 인자
def menu(wine, entree, dessert):
    return {'wine': wine, 'entree': entree, 'dessert': dessert}
print(menu(entree='beef', dessert='bagel', wine='bordeaux'))
print(menu('frontenac', dessert='flan', entree='fish'))
```

```python
# 기본 매개변수 값 정하기
def menu(wine, entree, dessert='pudding'):
    return {'wine': wine, 'entree': entree, 'dessert': dessert}
print(menu('chardonnay', 'chicken'))
print(menu('dunkelfelder', 'duck', 'doughnut'))
```

```python
def menu(price, wine='chardonnay', entree='chicken', dessert='pudding'):
    return {'price': price, 'wine': wine, 'entree': entree, 'dessert': dessert}
print(menu(100))
print(menu(price=100))
print(menu(price=120, entree='beef'))
print(menu(dessert='bagel', price=110))
print(menu('eighty', 'saint-pierre', 'fish'))
print(menu('hundred', wine='saint-pierre'))
```

```python
# 불가능한 경우
menu() # price에 인자가 할당이 안됨
menu(price=100, ‘saint-pierre’) # 키워드 인자 할당 후, 위치 인자
menu(100, price=120) # 같은 변수에 두 번 할당
menu(main=‘cream pasta’) # 정의되지 않은 매개변수
```

```python
# docstring
def echo(anything):
    'echo returns its input argument'
    return anything
```

```python
def print_if_true(thing, check):
    """
    Prints the first argument if a second argument is true.
    The operation is:
        1. Check whether the *second* argument is true.
        2. If it is, print the *first* argument.
    """
    if check:
        print(thing)
help(print_if_true)
```
